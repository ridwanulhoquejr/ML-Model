```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Base_image.jpg")
template = cv2.imread('/content/drive/MyDrive/Colab Notebooks/template_image.jpg')
(tH, tW) = template.shape[:2]

print("Test Image")
cv2_imshow(image)
print("template image")
cv2_imshow(template)
```

```python
imageGray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
templateGray = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)

print("[INFO]performing template matching...")
result = cv2.matchTemplate(imageGray, templateGray,	cv2.TM_SQDIFF_NORMED)
```

```python
imageGray.shape, templateGray.shape
```

```python
threshold = 0.1

(yCoords, xCoords) = np.where(threshold >= result)
clone = image.copy()
print("matched locationS")

for (x, y) in zip(xCoords, yCoords):
	cv2.rectangle(clone, (x, y), (x + tW, y + tH), (255, 0, 0))
cv2_imshow(clone)
```
